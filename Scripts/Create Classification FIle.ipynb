```python
import pandas as pd
import numpy as np

df = pd.read_csv("final_integrated_model1.csv").copy()

# Create violation rate
df["violation_rate"] = np.where(
    df["TOTAL_TRAFFIC"] > 0,
    df["TOTAL_VIOLATIONS"] / df["TOTAL_TRAFFIC"],
    0
)

# Define high-violation months as top 25%
threshold = df["violation_rate"].quantile(0.75)
df["HighViolation"] = (df["violation_rate"] >= threshold).astype(int)

print("Threshold:", threshold)
print(df["HighViolation"].value_counts())

# Keep only clean predictors + target
df_model2 = df[[
    "YEAR", "MONTH", "Facility", "Direction",
    "Avg_Speed", "Freeflow", "Delta",
    "PABT_Bus_Volume", "PABT_Passenger_Volume",
    "HighViolation"
]].copy()

df_model2.to_csv("final_integrated_model2_classification.csv", index=False)
df_model2.head()
```
